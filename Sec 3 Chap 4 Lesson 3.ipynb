## Lesson 3. Plot Histograms of Raster Values

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import rioxarray as rxr
import earthpy as et
```

```python
data_path = et.data.get_data('colorado-flood')
```

```python
sns.set(font_scale=1.5, style='whitegrid')
```

```python
lidar_dem_im = rxr.open_rasterio(os.path.join(data_path,
                                              'spatial',
                                              'boulder-leehill-rd',
                                              'pre-flood',
                                              'lidar',
                                              'pre_DTM.tif'))
lidar_dem_im.shape
```

**Raster Histogram**

```python
f, ax = plt.subplots(figsize=(12,6))

lidar_dem_im.plot.hist(ax=ax,
                       color='purple')
ax.set(title='Distribution of Lidar DEM Elevation Values',
       xlabel='Elevation (m)',
       ylabel='Frequency')
plt.show()
```

Masked the data

```python
lidar_dem_im = rxr.open_rasterio(os.path.join(data_path,
                                              'spatial',
                                              'boulder-leehill-rd',
                                              'pre-flood',
                                              'lidar',
                                              'pre_DTM.tif'),
                                 masked=True)
lidar_dem_im.shape
```

```python
f, ax = plt.subplots(figsize=(12,6))

lidar_dem_im.plot.hist(ax=ax,
                       color='purple',
                       bins=30)
ax.set(title='Distribution of Lidar DEM Elevation Values',
       xlabel='Elevation (m)',
       ylabel='Frequency')
plt.show()
```

**Customise the Histogram**

Specifying bins can be a good way to look for patterns in the data

```python
f, ax = plt.subplots(figsize=(12,6))

lidar_dem_im.plot.hist(ax=ax,
                       color='purple',
                       bins=[1600,1800,2000,2100])
ax.set(title='Distribution of Lidar DEM Elevation Values',
       xlabel='Elevation (m)',
       ylabel='Frequency')
plt.show()
```
